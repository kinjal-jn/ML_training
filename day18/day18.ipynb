#### graph functions
1. plot                        - simple line plot
2. plot.area                   - filled version of line plot
3. plot.bar / plot.barh        - bar graph
4. plot.box                    -box plot
5. plot.density                - Density line (smooth curved)
6. plot.hexbin                 - hexagon version of scatter with density
7. plot.hist                   - histogram
8. plot.kde                    - kernel density estimator (Density graph)
9. plot.pie                    - pie chart of categorical data
10. plot.scatter               - scatter graph

# Seaborn

```python
import seaborn as sns
```

### categories
1. categorical plots (hue can be used | comparision with categorical data)
    a. countplot (single argument)
    b. barplot (categorical + numerical data |calculates mean by default)
    c. boxplot
    d. violinplot
    e.swarmplot
2. Distribution plot
    a. dist plot
    b. displot
    c. joint plot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
3. Matrix plot
    a. heatmap
    b. clusterplor
    e. pairplot(same as above)
4. regression plot
    a. lmplot
    b. regplot
5. Grid plots
    a. pairGrid
    b. JointGrid
    c. FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.countplot(x='day', data=tips, hue='sex')
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day',y='total_bill', data=tips)
```

```python
sns.barplot(x='day',y='total_bill', data=tips, estimator=sum)
# estimator calculates whatever is passed to it.. sum avg etc
```

```python
sns.barplot(x='day',y='total_bill', data=tips, hue='time', estimator=np.mean)
```

```python
sns.swarmplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='time')
```
